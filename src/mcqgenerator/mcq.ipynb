```python
quiz_prompt = PromptTemplate(
    input_variables=["text", "number", "subject", "tone"],
    template="""


  



You are an expert MCQ creator.
Create exactly {number} multiple-choice questions about {subject}.
Use a {tone} tone.

STRICT FORMAT FOR EVERY QUESTION:
Q1. [Question here]
A) [Option]
B) [Option]
C) [Option]
D) [Option]
Correct Answer: A) [repeat the correct option]

---

Text:
{text}

Now generate exactly {number} questions following the format above strictly.
"""
)
```

```python
import os
from dotenv import load_dotenv
from groq import Groq

load_dotenv()

file_path = '/Users/erhiten6gmail.com/Downloads/Ai Overview.txt'

with open(file_path, 'r') as file:
    TEXT = file.read()

client = Groq(api_key=os.getenv("GROQ_API_KEY"))

formatted_prompt = quiz_prompt.format(
    text=TEXT,
    number=3,
    subject="machine learning",
    tone="simple"
)

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": formatted_prompt}],
    max_tokens=2048,
    temperature=0.7
)

print(response.choices[0].message.content)
```
